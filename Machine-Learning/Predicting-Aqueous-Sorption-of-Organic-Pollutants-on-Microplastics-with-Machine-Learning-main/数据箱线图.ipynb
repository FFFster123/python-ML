```python
import pandas as pd  
import seaborn as sns  
import matplotlib.pyplot as plt  
  
# 读取CSV文件  
df = pd.read_csv('DATA_1.csv')  
  
# 设置图形大小和比例  
plt.figure(figsize=(12, 6))  

# 设置统一的颜色  
box_color = '#BDB76B'  # 箱线图颜色  
hist_color = '#BDB76B'   # 直方图颜色  
kde_color = 'black'      # KDE曲线颜色  

# 创建第一个子图用于箱线图  
plt.subplot(1, 2, 1)  # 1行2列的子图，当前是第一个子图  
sns.boxplot(y='TN', data=df, palette=[box_color])  # 绘制箱线图  
# 计算中位数  
median = df['TN'].median()  
# 格式化中位数并添加标签  
median_label = f'Median: {median:.2f}'  
plt.text(0.5, 1.01 * max(df['TN']), median_label, ha='center', va='center')

plt.title('')  
plt.xlabel('TN')  
plt.ylabel('')  

# 创建第二个子图用于数据分布图  
ax2 = plt.subplot(1, 2, 2)  # 1行2列的子图网格中的第2个子图  
sns.histplot(df['TN'], kde=True, color=hist_color, ax=ax2, line_kws={'color': kde_color, 'linewidth': 2})  # 在子图上绘制直方图和核密度估计  
#sns.scatterplot(x=df.index, y='a', data=df, ax=ax2)  # 绘制数据点分布图
ax2.set_title('TN')  
ax2.set_xlabel('')  
ax2.set_ylabel('Frequency')  
  
# 调整子图之间的间距（如果需要）  
#plt.tight_layout()  
# 显示图形  
plt.show()
```
