```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (11, 11)  #set default figure size
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler

import matplotlib.pylab as plt

from sklearn.decomposition import PCA  
```

```python
df = pd.read_csv('MPs_data.csv', index_col=None, engine='c') 
print('original datasize:', df.shape)
df.dropna(inplace=True)
print('adjusted datasize:', df.shape)
print('data:')
df
```

```python
#数据标准化
scaler = StandardScaler()  
scaled_data = scaler.fit_transform(df)  

# 创建PCA对象，指定降维后的维度为9  
pca = PCA(n_components=9)  

# 对数据进行PCA降维  
pca_data = pca.fit_transform(scaled_data)
  
# 绘制PCA降维后的数据  
plt.figure(figsize=(8, 6))  
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=df['E-(C=n/V)'], edgecolor='none', alpha=0.5, cmap='viridis')  
plt.xlabel('PC1')  
plt.ylabel('PC2')  
plt.colorbar()  
plt.show()
```

```python
# 创建一个空的 DataFrame  
df = pd.DataFrame(columns=["PC{}".format(i) for i in range(1, pca_data.shape[1] + 1)])  
  
# 将 PCA 结果赋值给 DataFrame 的相应列  
for i, col in enumerate(df.columns):  
    df[col] = pca_data[:, i]  
df
```
